## Individuals from CIS countries
#### CIS countires: Armenia, Azerbaijan, Belarus, Kazakhstan, Kyrgyzstan, Moldova, Russia, Uzbekistan, Turkmenistan 

```python
import pandas as pd
```

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
df_entities = pd.read_csv('panama_papers.nodes.entity.csv',low_memory=False)
df_individuals = pd.read_csv('panama_papers.nodes.officer.csv',low_memory=False)
df_intermediaries = pd.read_csv('panama_papers.nodes.intermediary.csv',low_memory=False)
df_relationships = pd.read_csv('panama_papers.edges.csv',low_memory=False)
df_addresses = pd.read_csv('panama_papers.nodes.address.csv',low_memory=False)
```

```python
df_merge = pd.merge(df_relationships, df_addresses, left_on='END_ID', right_on='node_id')
```

```python
df_merge.drop(['node_id', 'sourceID_y'], axis= 1 , inplace= True)
```

```python
df_CIS_individuals = df_addresses[(df_addresses.country_codes == 'ARM')|
(df_addresses.country_codes == 'AZE')|
(df_addresses.country_codes == 'BLR')|
(df_addresses.country_codes == 'KAZ')|
(df_addresses.country_codes == 'KGZ')|
(df_addresses.country_codes == 'MDA')|
(df_addresses.country_codes == 'RUS')|
(df_addresses.country_codes == 'UZB')|
(df_addresses.country_codes == 'TKM')] 
df_CIS_individuals1 = df_CIS_individuals.sort_values(['country_codes','node_id'], ascending=[True, True])
```

```python
print(df_CIS_individuals1.shape)
```

```python
df1 = pd.merge(df_CIS_individuals1, df_relationships, left_on='node_id', right_on='END_ID')
```

```python
print(df1.shape)
```

```python
df1.drop(['node_id', 'sourceID_y','sourceID_x', 'valid_until_x', 'valid_until_y','note','start_date','end_date'], axis= 1 , inplace= True)
```

```python
df2 = pd.merge(df1, df_individuals, left_on='START_ID', right_on='node_id')
```

```python
print(df2.shape)
```

```python
df2.drop(['name_x', 'country_codes_y', 'countries_y','node_id','note','sourceID','valid_until'], axis= 1 , inplace= True)
```

```python
df_sorted1 = df2[['START_ID','name_y','TYPE','link','END_ID','address', 'country_codes_x','countries_x']]
```

```python
df_sorted1.rename(columns = {'START_ID':'IND_ID', 'name_y':'Name','END_ID':'ADR_ID','country_codes_x':'Country_codes','countries_x':'Countries'}, inplace = True)
```

```python
pd.concat([df_sorted1[col].astype(str).str.upper() for col in df_sorted1.columns], axis=1)
df_sorted2 = df_sorted1.drop_duplicates(['Name', 'address'])
print(df_sorted2.shape)
```

```python
df_Ent_cis = df_sorted2.loc[df_sorted2['Name']!='THE BEARER']
df_Ent_cis1 =df_Ent_cis.loc[df_sorted2['Name']!='BEARER']
print(df_Ent_cis1.shape)
```
