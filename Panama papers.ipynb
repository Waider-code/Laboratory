```python
import pandas as pd
```

```python
pd.set_option('display.max_rows',None)
```

```python
df_entities = pd.read_csv('panama_papers.nodes.entity.csv',low_memory=False)
```

```python
print(df_entities.shape)
```

## List of entities related to Estonia, Latvia, Lithuania from Panama papers database

```python
df_entities_panamapapres = df_entities[(df_entities.country_codes == 'LVA')|
(df_entities.country_codes == 'LTU')|
(df_entities.country_codes == 'EST')] 
df_entities_panamapapres_new = df_entities_panamapapres.sort_values(['country_codes','node_id'], ascending=[True, True])
```

```python
print(df_entities_panamapapres_new.shape)
#Total rows are 2293
```

```python
df_entities_panamapapres_new.head(15)
# list of entities related to Estonia, Latvia and Lithuania sorted by country and id
```

```python
df_Company_related_to_CIS = df_entities[(df_entities.country_codes == 'ARM')|
(df_entities.country_codes == 'AZE')|
(df_entities.country_codes == 'BLR')|
(df_entities.country_codes == 'KAZ')|
(df_entities.country_codes == 'KGZ')|
(df_entities.country_codes == 'MDA')|
(df_entities.country_codes == 'RUS')|
(df_entities.country_codes == 'UZB')|
(df_entities.country_codes == 'TKM')] 
df_CIS_companies = df_Company_related_to_CIS.loc[df_Company_related_to_CIS['name']!='THE BEARER']
df_CIS_companies_new = df_CIS_companies.loc[df_CIS_companies['name']!='BEARER']
df_CIS_related_companies = df_CIS_companies_new.sort_values(['country_codes','node_id'], ascending=[True, True])
```

```python
print(df_CIS_related_companies.shape)
#Total rows are 4232
```

```python
df_CIS_related_companies.head(100)
```

## List of intermidiaries related to Baltic and CIS countries from Panama papers database
#### Intermidiaries are Agents that help beneficiaries setup offshore companies

```python
df_intermidiaries = pd.read_csv('panama_papers.nodes.intermediary.csv',low_memory=False)
```

```python
print(df_intermidiaries.shape)
```

```python
df_intermidiaries_new = df_intermidiaries[(df_intermidiaries.country_codes == 'ARM')|
(df_intermidiaries.country_codes == 'AZE')|
(df_intermidiaries.country_codes == 'BLR')|
(df_intermidiaries.country_codes == 'KAZ')|
(df_intermidiaries.country_codes == 'KGZ')|
(df_intermidiaries.country_codes == 'MDA')|
(df_intermidiaries.country_codes == 'RUS')|
(df_intermidiaries.country_codes == 'UZB')|
(df_intermidiaries.country_codes == 'TKM')|
(df_intermidiaries.country_codes == 'LVA')|
(df_intermidiaries.country_codes == 'LTU')|
(df_intermidiaries.country_codes == 'EST')]
df_Intermediaries_CIS_Baltic = df_intermidiaries_new.sort_values(['country_codes','node_id'], ascending=[True, True])
```

```python
print(df_Intermediaries_CIS_Baltic.shape)
```
